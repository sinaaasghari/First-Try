```python
import numpy as np
import pandas as pd
```

```python
a='R'
b='98'
tr98=[]
for file in pd.ExcelFile('data_sets/project1/'+a+b+'.xlsx').sheet_names:
    tr98.append(pd.read_excel('data_sets/project1/'+a+b+'.xlsx',sheet_name=file))
```

```python
def object_to_int(x):
    x = str(x)
    counter = 0
    tf = False
    x = x.replace(" ", "")
    if x == 'nan':
        return np.nan
    else:
        for i in range(0,len(x)):
            if x[i] == '0' and tf == False:
                pass
            else:
                tf = True
                counter = 10 * counter
                counter += int(x[i])
    return counter
```
