```python
def read_data(a,b):
    temp=[]
    for file in pd.ExcelFile('data/'+a+b+'.xlsx').sheet_names:
        temp.append(pd.read_excel('data/'+a+b+'.xlsx',sheet_name=file))
    df=temp[0]
    #df=df.merge(temp[1],on='Address',how='left')
    df=df.merge(temp[2],on='Address',how='left')
    df=df.merge(temp[3].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Cost1'}),on='Address',how='left')
    df=df.merge(temp[4].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Cost2'}),on='Address',how='left')
    df=df.merge(temp[5].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Cost3'}),on='Address',how='left')
    df=df.merge(temp[6].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Cost4'}),on='Address',how='left')
    df=df.merge(temp[7].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Cost5'}),on='Address',how='left')
    df=df.merge(temp[8].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Cost6'}),on='Address',how='left')
    df=df.merge(temp[9].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Cost7'}),on='Address',how='left')
    df=df.merge(temp[10].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Cost8'}),on='Address',how='left')
    df=df.merge(temp[11].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Cost9'}),on='Address',how='left')
    df=df.merge(temp[13].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Cost11'}),on='Address',how='left')
    df=df.merge(temp[14].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Cost12'}),on='Address',how='left')
    df=df.merge(temp[15].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Cost13'}),on='Address',how='left')
    df=df.merge(temp[16].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Cost14'}),on='Address',how='left')
    #house  price 1
    df=df.merge(temp[6][temp[6]['code'].apply(lambda x:x//1000==41)].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'HousePrice1'}),on='Address',how='left')
    #house price 2
    df=df.merge(temp[6][temp[6]['code'].apply(lambda x:x//1000==42)].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'HousePrice2'}),on='Address',how='left')
    # fuel
    df=df.merge(temp[6][temp[6]['code'].apply(lambda x:x//1000==45)].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Fuel'}),on='Address',how='left')
    #car fuel
    df=df.merge(temp[9][temp[9]['code'].apply(lambda x:x//100==722)].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Car_Fuel'}),on='Address',how='left')
    #benzin
    df=df.merge(temp[9][temp[9]['code'].apply(lambda x:x==72211)].groupby('Address').agg({'value':'sum'}).rename(columns={'value':'Benzin'}),on='Address',how='left')
    return df
```

```python
df=[]
for a in ['R','U']:
    temp=[]
    for b in ['98','99','1400','1401']:
        print(a,b)
        x=read_data(a,b)
        x['year']=b
        temp.append(x)
    x=pd.concat(temp)
    x['state']=a
    df.append(x)
df=pd.concat(df)
```
