```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_household_distribution(data):
    plt.figure(figsize=(14, 10))  # تنظیم اندازه شکل

    # میزان تحصیلات
    plt.subplot(2, 2, 1)
    sns.histplot(data['Degree'], kde=False, bins=20)
    plt.title('Distribution of Education Level')
    plt.xlabel('Education Level')
    plt.ylabel('Count')

    # سن
    plt.subplot(2, 2, 2)
    sns.histplot(data['Age'], kde=False, bins=20)
    plt.title('Distribution of Age')
    plt.xlabel('Age')
    plt.ylabel('Count')

    # بستگی با سرپرست
    plt.subplot(2, 2, 3)
    sns.countplot(y=data['Relation'], order=data['Relation'].value_counts().index)
    plt.title('Distribution of Relationship with Head of Household')
    plt.xlabel('Count')
    plt.ylabel('Relationship')

    # وضعیت فعالیت
    plt.subplot(2, 2, 4)
    sns.countplot(y=data['Occupationalst'], order=data['Occupationalst'].value_counts().index)
    plt.title('Distribution of Employment Status')
    plt.xlabel('Count')
    plt.ylabel('Employment Status')

    plt.tight_layout()
    plt.show()
```

```python
# یک نمونه دیتافریم بدون داده واقعی
sample_data = {
    'Degree': [],  # لیست مقادیر میزان تحصیلات
    'Age': [],  # لیست مقادیر سن
    'Relation': [],  # لیست مقادیر بستگی با سرپرست
    'Occupationalst': []  # لیست مقادیر وضعیت فعالیت
}
data = pd.DataFrame(sample_data)
```

```python
# فراخوانی تابع برای رسم نمودارها
plot_household_distribution(data)
```
