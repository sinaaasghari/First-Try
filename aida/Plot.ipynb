```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_household_distribution(data):
    plt.figure(figsize=(14, 10))  # تنظیم اندازه شکل
    
    # میزان تحصیلات
    plt.subplot(2, 2, 1)
    sns.histplot(data['degree'], kde=False, bins=10, color='skyblue', binwidth=1)
    plt.title('Distribution of Education Level', fontsize=14)
    plt.xlabel('Education Level', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    
    # سن
    plt.subplot(2, 2, 2)
    sns.histplot(data['age'], kde=False, bins=20, color='lightgreen')
    plt.title('Distribution of Age', fontsize=14)
    plt.xlabel('Age', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    
    # بستگی با سرپرست
    plt.subplot(2, 2, 3)
    sns.countplot(y=data['relation'], hue=data['relation'], order=data['relation'].value_counts().index, palette='viridis', dodge=False)
    plt.title('Distribution of Relationship with Head of Household', fontsize=14)
    plt.xlabel('Count', fontsize=12)
    plt.ylabel('Relationship', fontsize=12)
    plt.legend([],[], frameon=False)  # حذف لِجِند
    
    # وضعیت فعالیت
    plt.subplot(2, 2, 4)
    sns.countplot(y=data['occupationalst'], hue=data['occupationalst'], order=data['occupationalst'].value_counts().index, palette='plasma', dodge=False)
    plt.title('Distribution of Employment Status', fontsize=14)
    plt.xlabel('Count', fontsize=12)
    plt.ylabel('Employment Status', fontsize=12)
    plt.legend([],[], frameon=False)  # حذف لِجِند
    
    plt.tight_layout()  # تنظیم چیدمان
    plt.show()
```

```python
# یک نمونه دیتافریم بدون داده واقعی
sample_data = {
    'degree': [],  # لیست مقادیر میزان تحصیلات
    'age': [],  # لیست مقادیر سن
    'relation': [],  # لیست مقادیر بستگی با سرپرست
    'occupationalst': []  # لیست مقادیر وضعیت فعالیت
}
data = pd.DataFrame(sample_data)
```

```python
# وارد کردن دیتا
data = pd.read_csv('personal_informations.csv') 
print(data.columns)
```

```python
# فراخوانی تابع برای رسم نمودارها
plot_household_distribution(data)
```
