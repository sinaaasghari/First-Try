```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore', category=UserWarning, append=True)
```

```python
data = pd.read_csv('df.csv')
print("Columns in the original file:", data.columns)
```

```python
data = data.rename(columns={
    'Cost3': 'clothing_shoes_expense',
    'Cost1': 'food_expense',
    'Cost4': 'housing_water_sewage_fuel_light_expense',
    'Cost6': 'health_expense'
})
print("Columns after renaming:", data.columns)
```

```python
columns_of_interest = ['clothing_shoes_expense', 'food_expense', 'housing_water_sewage_fuel_light_expense', 'health_expense']
data_filtered = data[columns_of_interest]
correlation_matrix = data_filtered.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Household Expenses', fontsize=16 , color='green')
plt.show()
```
